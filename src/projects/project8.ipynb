```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../../data/penguins.csv", index_col=0, sep=";")
df
```

```python
import math
avg_length = math.floor(df["Culmen Length (mm)"].mean())
avg_depth = math.floor(df["Culmen Depth (mm)"].mean())

print(f"The avg penguin's culmen length = {avg_length}mm and depth = {avg_depth}mm")

fig, axs = plt.subplots(ncols=2)

sns.histplot(data=df, x="Culmen Length (mm)", kde=True, color="purple", ax=axs[0])
sns.histplot(data=df, x="Culmen Depth (mm)", kde=True, ax=axs[1])
```

### Question 2

```python
fig, axs = plt.subplots(ncols=2)
sns.histplot(data=df, x="Culmen Length (mm)", kde=True, ax=axs[0], hue="Species")
sns.histplot(data=df, x="Culmen Depth (mm)", kde=True, ax=axs[1], hue="Species")

plt.show()
```

### Question 3

```python
lengths = df.groupby(["Species"])["Culmen Length (mm)"].mean()
depths = df.groupby(["Species"])["Culmen Depth (mm)"].mean()

print(lengths)
print(depths)
```

Based on the data above, it is evident that the Gentoos have long (47.6mm) and slim (14.9mm) beaks, while the Adelies have short (38.8mm) and wide (18.3mm) beaks.
