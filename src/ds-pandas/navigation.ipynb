```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../../data/panda_sector.csv")
```

```python
sns.scatterplot(data=df, x="x", y="y", hue='class', size='size')
plt.xlim(-10, 110)
plt.ylim(-10, 110)
plt.xlabel("x coordinate $\sqrt{warps}$", color="black")
plt.ylabel("y coordinate $\sqrt{warps}$", color="black")
```

```python
sns.scatterplot(data=df, x="x", y="y", hue="class", size="size")
plt.grid()
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')

plt.grid()
plt.annotate("black hole",
             xy=(0.6, 0.6),
             xycoords='data',
             xytext=(-90, -50),
             textcoords='offset points',
             fontsize=12,
             color="red",
             arrowprops={
                 'arrowstyle': "->",
                 'connectionstyle': "arc3,rad=.2",
                 'color': "red"
             })
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
plt.title("Panda sector x/y", color="black")
```
